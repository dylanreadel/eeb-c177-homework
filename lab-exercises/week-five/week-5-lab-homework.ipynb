# Part 1: Figures

```python
import numpy as np

data = np.loadtxt(fname = '../week-four/inflammation-01.csv', delimiter = ',')
```

```python
# 1.

import matplotlib.pyplot as plt

std_infl = np.std(data, axis = 0)
std_plot = plt.plot(std_infl)
plt.xlabel('time (days)')
plt.ylabel('standard deviation inflammation score')
plt.title('Standard Deviation of Inflammation Scores for all Patients')
plt.show()
```

```python
# 2.

fig = plt.figure(figsize=(5.0, 10.0))

fig.suptitle('Standard Deviation, Max, and Min Inflammation by Day', va='bottom', fontsize=12)

sbplot1 = fig.add_subplot(3, 1, 1)
sbplot2 = fig.add_subplot(3, 1, 2)
sbplot3 = fig.add_subplot(3, 1, 3)

sbplot1.set_ylabel('standard deviation')
sbplot1.set_xlabel('time (days)')
sbplot1.plot(np.std(data, axis=0))

sbplot2.set_ylabel('max')
sbplot2.set_xlabel('time (days)')
sbplot2.plot(np.max(data, axis=0))

sbplot3.set_ylabel('min')
sbplot3.set_xlabel('time (days)')
sbplot3.plot(np.min(data, axis=0))

fig.tight_layout()   # auto reformat the size of the plots

plt.show()
```

# Part 3: Lists

1. A list contains strings, integers, or floats separated by commas. The items are changeable and it allows duplicates.

2. To create a list, set a variable equal to items in square brackets. To index values, use `print(listname[index position])`

3. To change a value of an individual element, use the command: `listname[index position] = 'new value'`

4. To append a new value, use the command: `listname.append(new value)` or `listname.insert(index position, new value)`

```python
a = [1, 2, 3, 4]
```

```python
print(a[2])
```

```python
a[1] = 6
print(a)
```

```python
a.append(7)
print(a)
```

```python
a.insert(2, 15)
print(a)
```

# Part 4: Conditionals

1. If this code were ran, 'C' would be printed because 4 is less than 5 which is equal to the last `elif` statement.

```python
string = str(input('Input a word or sentence: '))

vowels = 0

for v in string:
    if (v == 'a' or
        v == 'e' or
        v == 'i' or
        v == 'o' or
        v == 'u' or
        v == 'A' or
        v == 'E' or
        v == 'I' or
        v == 'O' or
        v == 'U'):
        vowels = vowels + 1
    print("The number of vowels in the string are:", vowels)
```
