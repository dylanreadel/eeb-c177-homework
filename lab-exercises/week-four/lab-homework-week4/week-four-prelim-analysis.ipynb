```python
# importing pandas for reading csv file with strings

import pandas as pd

# reading csv file for nestling fledging data
# making the date column the index column for plotting purposes

fledging_data = pd.read_csv('Pietz-fledging-time-data.csv', 
                            parse_dates=['Date_of_fledge'], 
                            index_col='Date_of_fledge')

fledging_data['Fledge_time'] = pd.to_datetime(fledging_data['Fledge_time'])
```
